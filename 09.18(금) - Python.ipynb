# Pandas 분석기능

데이터 요약 기능
- Groupby : 그룹별로 데이터를 요약
    - splitt(그룹객체)
    - apply
    - combine
    - Aggregation (집계함수: sum, avg, var)
    - filltering
- pivot       : 엑셀 타입 데이터. 축을 지정하고 값을 표현.
- pivot_table : Aggregation을 사용하여 변화된 값을 계층적 인덱스로 표현. 행렬(Key 할당 + Numpy구조)
- cross_tab : 범주형 변수
- 시간과 시계열 분석(시계열 데이터의 전처리) : 시계열 분석.
    - 정상성 <> 비정상성 (정상화 - 미분, power, sqrt, log ...)
    - 고주파를 저주파수로 변경(이동평균법)
    - ARMA(Auto Regression Moving Average)
    - ARIMA(Integrated) : 비정상성을 띄우더라도 분석이 가능하도록.
    - statsmodels

```python
df=pd.DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'],
                 'key2'   : ['one', 'two', 'one', 'two', 'one'],
                 'data1'  : np.random.randn(5),
                 'data2'  : np.random.randn(5)})
print(df)

gp = df['data1'].groupby(df['key1'])
print('\n그룹평균 = \n', gp.mean())

gp = df['data1'].groupby([df['key1'], df['key2']])
print('\n그룹평균 = \n', gp.mean())
```

```python
df = pd.DataFrame({'key1':['영업부', '영업부', '관리부', '관리부', '영업부'], 
                   'key2':['남', '여', '남', '여', '남'],
                   'data1':np.random.randn(5),
                   'data2':np.random.randn(5)})

# 부서별 평균 출력
gp = df.groupby(df['key1'])
print('\부서별 평균 = \n',gp.mean())

# 부서별 성별 평균 출력
gp = df.groupby([df['key1'], df['key2']])
print('\부서별 평균 = \n',gp.mean())

gp_unstack = gp.mean().unstack()
gp_unstack.stack()
```

```python
flights = pd.read_csv('flights.csv')
flights

#월, 일, 주, 항공사, 출발공항, 도착공항, 예정출발시간,
#출발지연시간, 운항시간, 운항거리, 예정도착시간, 도착지연시간,
#우회여부, 취소여부
```

```python
# 항공사별로 평균지연시간 출력
flights.groupby('AIRLINE')['ARR_DELAY'].mean().head()

# 항공사별로 주마다 비행을 취소한 합계
flights.groupby(['AIRLINE', 'WEEKDAY'])['CANCELLED'].sum().head(10)

# 항공사별로 주마다 비행을 취소하거나 우회한 합계와 평균
flights.groupby(['AIRLINE', 'WEEKDAY'])['CANCELLED', 'DIVERTED'].agg(['sum', 'mean']).head(98)

# 
group_cols = ['ORG_AIR', 'DEST_AIR']
agg_dict = {'CANCELLED':['sum', 'mean', 'size'],
            'AIR_TIME':['mean', 'var']}
flights.groupby(group_cols).agg(agg_dict).head(20)

# 항공사별 주별로 거리의 합계와 평균, 도착지연에 대해 최솟값과 최대값 출력
gp = ['AIRLINE', 'WEEKDAY']
ad = {'DIST':['sum', 'mean'],
      'ARR_DELAY':['min', 'max']}
airline_info = flights.groupby(gp).agg(ad).head(20)

# 항공사만 출력
airline_info.index.get_level_values(0)
level0 = airline_info.columns.get_level_values(0)
level1 = airline_info.columns.get_level_values(1)
airline_info.columns = level0 + '_' + level1
airline_info.head(3)
airline_info.reset_index().head()

flights.groupby(['AIRLINE'], as_index=False, sort=False)['DIST'].agg('mean').round(0).head(20)


# flights.groupby('AIRLINE')['ARR_DELAY'].agg('mean').values
# flights.groupby('AIRLINE')['ARR_DELAY'].agg('mean').keys()
# flights.groupby('AIRLINE').agg({'ARR_DELAY':'mean'})
# flights.groupby('AIRLINE')['ARR_DELAY'].agg(np.mean)
# flights.groupby('AIRLINE')[['ARR_DELAY', 'DEP_DELAY']].mean()
```

```python
df = pd.DataFrame({'A' : ['일반', '일반', '특수', '예외'] * 6,
                   'B' : ['서울', '대구', '부산'] * 8,
                   'C' : ['건조', '건조', '건조', '우기', '우기', '우기'] * 4,
                   'D' : np.random.randn(24),
                   'E' : np.random.randn(24)})
pd.crosstab(df.A, df.B, margins=True)
pd.crosstab(df.A, df.C, margins=True)
pd.crosstab([df.A, df.B], df.C, margins=True)
pd.crosstab(df.A, df.B).apply(lambda x: x/x.sum(), axis=1)
```
