```python
class Parent:
    parentAttr = 100
    def __init(self, name):
        self.name = name
        print("부모 생성자")
    def parentMethod(self):
        print("부모 메써드")
    def setAttr(self, attr):
        Parent.parentAttr = attr
    def getAttr(self):
        print("부모속성 : ", Parent.parentAttr)

class Child(Parent):
    def __init(self):
        print("자식 생성자")
    def childMethod(self):
        print("자식 메써드")
    def print_attribute(self):
        self.setAttr(500)
        print("자식에서 호출된 부모 속성 : ", Parent.parentAttr)
        
c = Child()
c.childMethod()
c.parentMethod()
c.setAttr(200)
c.getAttr()
c.print_attribute()
c.parentAttr
```

```python
class Animal:
    def Name(self):
        pass
    def Sleep(self):
        print("Sleep")
    def MakeNoise(self):
        pass
    
class Dog(Animal):
    def Name(self):
        print("강아지")
    def MakeNoise(self):
        print("멍멍")

class Cat(Animal):
    def Name(self):
        print("고양이")
    def MakeNoise(self):
        print("애옹")
        
class Lion(Animal):
    def Name(self):
        print("사자")
    def MakeNoise(self):
        print("커어어어흐으응")
        
class TestAnimals:
    def PrintName(self, animal):
        animal.Name()
    def GotoSleep(self, animal):
        animal.Sleep()
    def MakeNoise(self, animal):
        animal.MakeNoise()

testAnimals = TestAnimals()
dog = Dog()
cat = Cat()
lion = Lion()

testAnimals.PrintName(dog)
testAnimals.GotoSleep(dog)
testAnimals.MakeNoise(dog)
print()
testAnimals.PrintName(cat)
testAnimals.GotoSleep(cat)
testAnimals.MakeNoise(cat)
print()
testAnimals.PrintName(lion)
testAnimals.GotoSleep(lion)
testAnimals.MakeNoise(lion)
print()

nam = []
nam.append(dog)
nam.append(cat)
nam.append(lion)
for ma in nam:
    testAnimals.PrintName(ma)
    testAnimals.GotoSleep(ma)
    testAnimals.MakeNoise(ma)
```

# Shape를 상속받은 Triangle(밑변, 높이) 와 Rectangle(두 점)
# Circle(센터와 반지름) 클래스 작성
##### Point Class : x, y를 관리. -> 좌표를 표현 -> 포함의 개념()
##### Shape : 면적을 계산하는 함수, 둘레를 계산하는 함수 -> 부모 클래스 -> 상속의 개념
- 포함과 상속을 이용하여 구성.

```python
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Shape:
    def __init__(self):
        print("부모 클래스")
    def area(self):
        print("면적")
    def perimeter(self):
        print("둘레")
        
class Triangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        print("삼각형의 넓이는 = ", self.width*self.height/2)
    def perimeter(self):
        print("삼각형의 둘레는 = ", self.width + math.sqrt(math.pow(self.width/2, 2)) + math.pow(self.height, 2)*2)
    
class Rectangle(Shape):
    def __init__(self, po1, po2):
        self.po1 = po1
        self.po2 = po2
    def area(self):
        print("사각형의 넓이는 = ", abs(self.po1.x - self.po2.x) * abs(self.po1.y - self.po2.y))
    def perimeter(self):
        print("사각형의 둘레는 = ", abs(self.po1.x - self.po2.x)*2 + abs(self.po1.y - self.po2.y)*2)
        
class Circle(Shape):
    def __init__(self, po1, r):
        self.po1 = po1
        self.r = r
    def area(self):
        print("원의 넓이는 = ", math.pi * math.pow(self.r, 2))
    def perimeter(self):
        print("원의 둘레는 = ", 2 * math.pi * self.r)
```

```python
import random
shape_list = []
for i in range(10):
    if i%3 == 0:
        tri = Triangle(10, 20)
        shape_list.append(tri)
    elif i%3 == 1:
        rect = Rectangle(Point(10, 20), Point(20, 30))
        shape_list.append(rect)
    else:
        r = random.randint(0, 10)
        ci = Circle(Point(10, 20), 5)
        shape_list.append(ci)

for i in range(10):
    shape_list[i].area()
    shape_list[i].perimeter()
```

```python
class TwoWayDict(dict):
    def __delitem__(self, key):
        value = super().pop(key)
        super().pop(value, None)
        
    def __setitem__(self, key, value): # 용도를 변경, 대입방식으로 진행될 때 처리되는 함수
        if key in self:
            del self[self[key]]
        if value in self:
            del self[value]
        super().__setitem__(key, value) # 부모 클래스
        super().__setitem__(value, key)
        
    def __repr__(self):
        return f'{type(self).__name__}({super().__repr__()})'
    
    
d = TwoWayDict()
d[3] = 8
print(d)
d[7] = 6
print(d)
```

# 학교 관리 프로그램을 개발한 업체에서
# 학교마다 다양한 평가방법을 반영할 수 있도록 프로그램 개발.
- 평가 방법 : 학점(A,B,C,D,E,F)
- 합격 / 불합격 여부로 평가.
- 수우미양가 여부로 평가.

##### 위 3개의 방법으로 코드 수정 없이 작동시키는 프로그램 구현하시오.

```python
class InputScore:
    def __init__(self, score):
        self.score = score
    
class printGrade:
    def __init__(self):
        print("부모 클래스")
    def printGradeSuper(self):
        print("평가방법에 따라 출력")
        
class GradeSix(printGrade):
    def __init__(self, score):
        self.score = score
    def printGradeSuper(self, score):
        if score >= 90:
            print("성적은 = ", self.score, ", 등급은 : A")
        elif score >= 80:
            print("성적은 = ", self.score, ", 등급은 : B")
        elif score >= 70:
            print("성적은 = ", self.score, ", 등급은 : C")
        elif score >= 60:
            print("성적은 = ", self.score, ", 등급은 : D")
        elif score >= 50:
            print("성적은 = ", self.score, ", 등급은 : E")
        else:
            print("성적은 = ", self.score, ", 등급은 : F")
            
class PassOrNot(printGrade):
    def __init__(self, score):
        self.score = score
    def printGradeSuper(self, score):
        if score >= 60:
            print("성적은 = ", self.score, ", 합격여부 : PASS")
        else:
            print("성적은 = ", self.score, ", 합격여부 : Non_PASS")
            
class GradeFive(printGrade):
    def __init__(self, score):
        self.score = score
    def printGradeSuper(self, score):
        if score >= 90:
            print("성적은 = ", self.score, ", 등급은 : 수")
        elif score >= 80:
            print("성적은 = ", self.score, ", 등급은 : 우")
        elif score >= 70:
            print("성적은 = ", self.score, ", 등급은 : 미")
        elif score >= 60:
            print("성적은 = ", self.score, ", 등급은 : 양")
        else:
            print("성적은 = ", self.score, ", 등급은 : 가")
    
```

```python
num = int(input("성적을 입력하시오 : "))
while 1:
    choice = input("[ 1: ABCDEF \t 2: 합격여부 \t 3: 수우미양가 \t 4: 종료 ] => ")
    if choice == "1":
        six = GradeSix(num)
        six.printGradeSuper(num)
    elif choice == "2":
        PON = PassOrNot(num)
        PON.printGradeSuper(num)
    elif choice == "3":
        fiv = GradeFive(num)
        fiv.printGradeSuper(num)
    else:
        print("프로그램 종료")
        break


# printGrade_List = []
# printGrade_List.append(six)
# printGrade_List.append(PON)
# printGrade_List.append(fiv)
```

# 시계 클래스

```python
class Clock(object):
    def __init__(self, hours=0, minutes=0, seconds=0):
        self.__hours = hours
        self.__minutes = minutes
        self.__seconds = seconds
    def set(self, hours, minutes, seconds =0):
        self.__hours = hours
        self.__minutes = minutes
        self.__seconds = seconds
    def tick(self):
        if self.__seconds == 59:
            self.__seconds = 0
            if self.__minutes == 59:
                self.__minutes = 0
                if self.__hours == 23:
                    self.__hours = 0
                else:
                    self.__hours +=1
            else:
                self.__minutes +=1
        else:
            self.__seconds +=1
    def check(self):
        if self.__hours == 0 and self.__seconds == 0 and self.__minutes == 0:
            return True
    def __str__(self):
        return "%2d:%2d:%2d" % (self.__hours, self.__minutes, self.__seconds)
    
x = Clock()
print(x)
for i in range(23*60*60):
    x.tick()
print(x)
```

# 달력 클래스

```python
class Calendar(object):
    months = (31,28,31,30,31,30,31,31,30,31,30,31)
    def __init__(self, day=1, month=1, year=1900):
        self.__day = day
        self.__month = month
        self.__year = year
    def leapyear(self, y):
        if y%4:
            return 0
        else:
            if y%100:
                return 1
            else: 
                if y%400:
                    return 0
                else:
                    return 1
    def set(self, day, month, year):
        self.__day = day
        self.__month = month
        self.__year = year
    def get():
        return(self, self.__day, self.__month, self.__year)
    def advance(self):
        month = Calendar.months
        max_days = month[self.__month-1]
        if self.__month == 2:
            max_days += self.leapyear(self.__year)
        if self.__day ==  max_days:
            self.__day = 1
            if (self.__month == 12):
                self.__month = 1
                self.__year += 1
            else:
                self.__month += 1
        else:
            self.__day += 1 
    def __str__(self):
        return str(self.__year) + "/" + str(self.__month) + "/" + str(self.__day)
    

    
x = Calendar()
print(x)
for u in range(365*2):
    x.advance()
    print(x)
```

# Clock 과 Calendar 클래스를 다중 상속하여
# 시간 및 캘린더가 동시에 작용하는 프로그램을 구현하시오

```python
class CalendarClock(Clock, Calendar):
    def __init__(self, year=2000, month=1, day=1, hours=23, minutes=59, seconds=40):
        Calendar.__init__(self, day, month, year)
        Clock.__init__(self, hours, minutes, seconds)
    def __str__(self):
        return Calendar.__str__(self) + ", " + Clock.__str__(self)
    def tick_advance(self):
        self.tick()
        if self.check():
            self.advance()
            
cacl = CalendarClock()
for u in range(20):
    cacl.tick_advance()
    print(cacl)
```
